```python
from sympy.solvers import solve
from sympy import Symbol
from sympy.plotting import plot
from sympy.plotting import plot_implicit
%matplotlib inline

x = Symbol('x')
y = Symbol('y')


expr = x**2 - x - 1
roots = solve(expr)
roots[0]**2
# print(len(roots))
# roots.count(0)
# roots.count
```

```python
p = plot(expr, (x, -4, 4), show=False, aspect_ratio=(1,1))
p.ylim=(-2, 2)
```

```python
p.show()
```

```python
p2 = plot(x + 1, (x, -4, 4), show=False, aspect_ratio=(1,1))
# p2.xlim = (-3, 3)
p2.ylim = (-2, 2)
p.extend(p2)
p.xlim = (-3, 3)
# from matplotlib import style
# style.use('dark_background')

p.show()
```
