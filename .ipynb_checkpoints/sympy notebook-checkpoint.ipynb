```python
from sympy.solvers import solve
from sympy import Symbol
from sympy.plotting import plot
from sympy.plotting import plot_implicit
x = Symbol('x')
y = Symbol('y')

expr = x**2 - x - 1
roots = solve(expr)
roots[0]**2
# print(len(roots))
# roots.count(0)
# roots.count
```

```python
p = plot(expr, (x, -3, 3), (y, 3, -2), show=False)
p.show()
```
